```python
import pandas as pd 
import numpy as np
import re
from zemberek import (
        TurkishSpellChecker,
    TurkishSentenceNormalizer,
    TurkishSentenceExtractor,
    TurkishMorphology,
    TurkishTokenizer
)
```

```python
data = pd.read_csv("corpus/texts.csv") # https://github.com/selimfirat/bilkent-turkish-writings-dataset
data1 = ''
data2 = ''
for i in range(20):
    data1 = data1 + data['text'][i] 
    data2 = data2 + data['text'][i+20] 
```

```python
data1 = re.sub('[“”’‘;.,:?!0-9]' , '' , data1)
data2 = re.sub('[“”’‘;.,:?!0-9]' , '' , data2)
```

```python
data1 = (data1)
data1 = data1.lower()
```

```python
data2 = (data2)
data2 = data2.lower()
```

```python
from nltk.tokenize import word_tokenize
```

```python
data1 = word_tokenize(data1)
```

```python
data2 = word_tokenize(data2)
```

```python
def by_value(item):
    return item[1]
```

```python
def count_n_grams(data, n , threshold):

    #n_gramları eklemek için dictionary
    n_grams = {}
    n_grams_ = {}
    n_gr = {}

    data = tuple(data)
    
    for i in range(len(data)): # satırın tamamını alıyoruz
        
        # i den i+n e kadar olan kısmı n-gram olarak alma
        n_gram = data[i:i+n]
            
        # n_grams ın içinde var mı kontrolü
        if n_gram in n_grams.keys(): 
            
            # Eğer varsa 1 artırıyoruz
            n_grams[n_gram] += 1
        else:
            # Yoksa 1 eşitleyerek eklemiş oluyoruz
            n_grams[n_gram] = 1
            
    len_ = len(n_grams)
                
    for k, v in n_grams.items():
        if v > threshold:
            n_grams_[k] = v 
            
    for k, v in sorted(n_grams_.items(), key=by_value):
        n_gr[k] = v
            
    return n_gr
```

```python
data1_uni = count_n_grams(data1, 1 ,1)
print(len(data1_uni))
```

```python
data2_uni = count_n_grams(data2, 1 ,1)
print(len(data2_uni))
```

```python
data1_bi = count_n_grams(data1, 2 ,1)
print(len(data1_bi))
```

```python
data2_bi = count_n_grams(data2, 2 ,1)
print(len(data2_bi))
```

```python
data1_tri = count_n_grams(data1, 3 ,1)
print(len(data1_tri))
```

```python
data2_tri = count_n_grams(data2, 3 ,1)
print(len(data2_tri))
```
