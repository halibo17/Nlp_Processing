# N-grams Corpus Processing

```python
import pandas as pd 
import numpy as np

data = pd.read_csv("corpus/texts.csv") # https://github.com/selimfirat/bilkent-turkish-writings-dataset
data = data['text'][0] + data['text'][1] + data['text'][2]
data
```

```python
from nltk.tokenize import word_tokenize
```

```python
data_ = []
data_ = word_tokenize(data.lower())
data_
```

```python
def count_n_grams(data, n , threshold):

    #n_gramları eklemek için dictionary
    n_grams = {}
    n_grams_ = {}

    data = tuple(data)
    
    for i in range(len(data)): # satırın tamamını alıyoruz
        
        # i den i+n e kadar olan kısmı n-gram olarak alma
        n_gram = data[i:i+n]
            
        # n_grams ın içinde var mı kontrolü
        if n_gram in n_grams.keys(): 
            
            # Eğer varsa 1 artırıyoruz
            n_grams[n_gram] += 1
        else:
            # Yoksa 1 eşitleyerek eklemiş oluyoruz
            n_grams[n_gram] = 1
    len_ = len(n_grams)
                
    for k, v in n_grams.items():
        if v > threshold:
            n_grams_[k] = v 
            
    return n_grams_
```

```python
print("Uni-gram:")
count_n_grams(data_, 1 ,5)
```

```python
print("Bi-gram:")
count_n_grams(data_, 2 ,5)
```

```python
print("Tri-gram:")
count_n_grams(data_, 3 ,3)
```
