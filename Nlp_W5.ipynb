```python
def by_value(item):
    return item[1]
```

```python
def count_n_grams(data, n , threshold):

    #n_gramları eklemek için dictionary
    n_grams = {}
    n_grams_ = {}
    n_gr = {}

    data = tuple(data)
    
    for i in range(len(data)): # satırın tamamını alıyoruz
        
        # i den i+n e kadar olan kısmı n-gram olarak alma
        n_gram = data[i:i+n]
            
        # n_grams ın içinde var mı kontrolü
        if n_gram in n_grams.keys(): 
            
            # Eğer varsa 1 artırıyoruz
            n_grams[n_gram] += 1
        else:
            # Yoksa 1 eşitleyerek eklemiş oluyoruz
            n_grams[n_gram] = 1
            
    len_ = len(n_grams)
                
    for k, v in n_grams.items():
        if v > threshold:
            n_grams_[k] = v 
            
    for k, v in sorted(n_grams_.items(), key=by_value):
        n_gr[k] = v
            
    return n_gr
```

```python
a = count_n_grams(data, 1 , 1)
```

```python
from sklearn.decomposition import TruncatedSVD

# SVD represent documents and terms in vectors 

svd_model = TruncatedSVD(n_components=1, algorithm='randomized', n_iter=100, random_state=122)
lsa = svd_model.fit_transform(X)
```
