```python
import time
import logging
import pandas as pd 
import numpy as np
import re
from zemberek import (
        TurkishSpellChecker,
    TurkishSentenceNormalizer,
    TurkishSentenceExtractor,
    TurkishMorphology,
    TurkishTokenizer
)
```

```python
data = pd.read_csv("corpus/texts.csv") # https://github.com/selimfirat/bilkent-turkish-writings-dataset
print(data['text'][0])
```

```python
data = (data['text'][0] )
data = data.lower()
```

```python
data = re.sub('[\r]' , '',(data) )  
```

```python
data = data.split()
```

```python
morphology = TurkishMorphology.create_with_defaults()
```

```python
data_ = []
for i in range(len(data)):
    result = morphology.analyze(data[i])
    print(result)
    print("\n")
    data_.append(result)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
tokens = nltk.word_tokenize(' '.join(data))
```

```python
print(nltk.pos_tag(tokens))
```
